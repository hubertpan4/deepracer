```python
curve = LineString([(0, 0), (1, 1), (2,4) ,(3,9)])
intersectingLine = LineString([(0,1), [4,1]])
nonIntersectingLine = LineString([(0,1), [3,11]])

expectedIntersections = curve.intersects(intersectingLine)
print(expectedIntersections)
```

```python
curve.intersects(nonIntersectingLine)
```

```python
# extract the x, y points of each sequence we want to plot
cx, cy = curve.xy
ix, iy = intersectingLine.xy
nx, ny = nonIntersectingLine.xy

plt.plot(cx, cy, label="curve")
plt.plot(ix, iy, label="intersectingLine")
plt.plot(nx, ny, label="nonIntersectingLine")
```

```python
leftCurve = curve.offset_curve(-1)
rightCurve = curve.offset_curve(1)

lcx, lcy = leftCurve.xy 
rcx, rcy = rightCurve.xy 

plt.plot(cx, cy, label="curve")
plt.plot(lcx, lcy, label="left")
plt.plot(rcx, rcy, label="right")
plt.xticks(range(0,10))
plt.yticks(range(0,10))
plt.legend()
```

```python
leftCurve = curve.parallel_offset(1, "left")
rightCurve = curve.parallel_offset(1, "right")

cx, cy = curve.xy
lcx, lcy = leftCurve.xy 
rcx, rcy = rightCurve.xy 

plt.plot(cx, cy, label="curve")
plt.plot(lcx, lcy, label="left")
plt.plot(rcx, rcy, label="right")
plt.xticks(range(0,10))
plt.yticks(range(0,10))
plt.legend()
```

```python
def spliceWithLoop(array: list, start: int, stop: int) -> list:
    arrayLen = len(array)
    if (start < arrayLen and stop < arrayLen):
        return array[start:stop]
    if (start < arrayLen and stop == arrayLen):
        return array[start:stop]
    if (start < arrayLen and stop > arrayLen):
        result = array[start:]
        stop = stop-arrayLen
        while stop > arrayLen:
            result = result + array
            stop = stop - arrayLen
        result = result + array[:stop]
        return result 
        
```

```python
print(spliceWithLoop([0,1,2,3,4,5], 0, 3))
print(spliceWithLoop([0,1,2,3,4,5], 3, 5))
print(spliceWithLoop([0,1,2,3,4,5], 3, 6))
print(spliceWithLoop([0,1,2,3,4,5], 3, 7))
```

```python
import matplotlib.pyplot as plt
from shapely.geometry import LineString
import numpy as np

# Example coordinates for a sine curve
x = np.arange(0, 6 * np.pi, 0.01)
y = np.sin(0.5*x) * 2
line = LineString(zip(x, y))

def calculate_curvature(line):
    coords = np.array(line.coords)
    dx = np.gradient(coords[:, 0])
    dy = np.gradient(coords[:, 1])
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(ddx * dy - dx * ddy) / (dx**2 + dy**2)**1.5
    return curvature

curvature = calculate_curvature(line)

# Plotting the line and its curvature
plt.plot(x, y, label='Line')
plt.plot(x, curvature, label='Curvature')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from shapely.geometry import LineString
import numpy as np

# Example coordinates for a sine curve
y = np.arange(0, 6 * np.pi, 0.01)
x = np.sin(1*x) * 2
line = LineString(zip(x, y))

def calculate_curvature(line):
    coords = np.array(line.coords)
    dx = np.gradient(coords[:, 0])
    dy = np.gradient(coords[:, 1])
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(ddx * dy - dx * ddy) / (dx**2 + dy**2)**1.5
    return curvature

curvature = calculate_curvature(line)

# Plotting the line and its curvature
plt.plot(x, y, label='Line')
plt.plot(x, curvature, label='Curvature')
plt.legend()
plt.show()
```
